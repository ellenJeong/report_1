```python
import urllib.request
import bs4
url = "https://news.naver.com"
html = urllib.request.urlopen(url)

bs_obj = bs4.BeautifulSoup(html, "html.parser")
newsnow_txarea = bs_obj.findAll("div", {"class" : "cjs_news_tw"})

for i in newsnow_txarea:
    div_tag=i.find("div")
    print(re.sub(r"[^\uAC00-\uD7A30-9a-zA-Z\s]", "", div_tag.text))
    p_tag=i.find("p")
    print(re.sub(r"[^\uAC00-\uD7A30-9a-zA-Z\s]", "", p_tag.text))
    print("===========================================================")
```
